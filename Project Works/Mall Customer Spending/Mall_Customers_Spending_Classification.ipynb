# Random Forest Classification

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, 1:-2].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

## Encoding data

### Encoding the Binary Independent Variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:,0] = le.fit_transform(X[:,0])
y= le.fit_transform(y)
```

```python
print(X)
```

```python
print(y)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[:, 1:3] = sc.fit_transform(X_train[:, 1:3])
X_test[:, 1:3] = sc.transform(X_test[:, 1:3])
```

```python
print(X_train)
```

```python
print(X_test)
```

## Training the Random Forest Classification model on the Training set

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 42,max_depth=5,class_weight='balanced')
classifier.fit(X_train, y_train)
```

## Predicting a new result

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
